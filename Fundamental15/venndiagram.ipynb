```python
import matplotlib.pyplot as plt
from matplotlib_venn import *
```

```python
fig = plt.figure()
fig.set_size_inches(10,8)

Num_Set_A = 103            # A 집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.
Num_Set_B = 77             # B 집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.
Num_Set_A_and_B = 5        # A와 B의 교집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.

v = venn2(subsets={'10': Num_Set_A, '01': Num_Set_B, '11': Num_Set_A_and_B}, set_labels = ('A', 'B'))
c = venn2_circles(subsets=(Num_Set_A, Num_Set_B, Num_Set_A_and_B), linestyle='solid') 
v.get_patch_by_id('10').set_alpha(0.8)
v.get_patch_by_id('10').set_color('w') 
v.get_patch_by_id('01').set_color('w') 
v.get_patch_by_id('11').set_color('yellow') 


v.get_label_by_id('10').set_text(' ') 
v.get_label_by_id('01').set_text(' ') 
v.get_label_by_id('11').set_text(' ') 

v.get_label_by_id('A').set_text('Set A') 
v.get_label_by_id('B').set_text('Set B')

# custom2 = venn2(subsets = (1,2,3), set_labels=('A','B'))

plt.gca().set_facecolor("white")
plt.gca().set_axis_on()
plt.title('First venndiagram')
plt.show()
```

venn2는 2개의 벤다이어그램을 만든다.  
10,01,11은 부분을, 라벨 A,B는 원 1개씩을 말한다.  
subset dict 넣으면 '10','01','11'를 key name 로 넣기, 다른 이름 넣으면 원이 생성되지 않는다.  
subset tuple로 넣으면 순서대로 3개(정해진거 없음)  
subset는 또 set 형식 2개를 가지는 리스트 또는 튜플를 넣을 수도 있다고 한다(교집합 알아서 만들어짐)  
출력시 만들어지는 숫자는 원소의 개수
set_labels = 벤다이어 그램 양쪽 이름, set_colors : 양쪽 색상 - 교집합은 알아서 색이 섞인다. 
venn2_circle : 테두리 원만 그리기
alpha : transparency(1.0 짙게, 0 얕게)  
set_text : 원 내부 내용 설정  
get_label_by_id: returns a subset label by one region ID; region ID is ‘10’, ‘01’ or ‘11’ for a 2-circle diagram and ‘100’, ‘110’, etc., for a 3-circle diagram.  10,01,11로 해서 그 subset label을 리턴  
get_label 에서 01,10,11로 하면 내부가 바뀌고 label(ex)A,B)를 쓰면 하단에 표시된다.  
get_patched_by_id : returns a patch by a region ID.  
patched와 label 마다 쓸 수 있는 method가 다르다.

plt.gca().set_facecolor("white") : facecolor를 흰색으로  
plt.gca().set_axis_on() : 축을 만들어줘서 facecolor가 들어가게 해줌

```python
fig = plt.figure()
fig.set_size_inches(10,8)

Num_Set_A = 13            # A 집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.
Num_Set_B = 8            # B 집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.
Num_Set_A_and_B = 5    # A와 B의 교집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.

v = venn2(subsets={'10': Num_Set_A, '01': Num_Set_B, '11': Num_Set_A_and_B}, set_labels = ('A', 'B')) 
c = venn2_circles(subsets=(Num_Set_A, Num_Set_B, Num_Set_A_and_B), linestyle='solid') 
v.get_patch_by_id('10').set_alpha(0.8) 

v.get_patch_by_id('10').set_color('lightgreen') 
v.get_patch_by_id('01').set_color('lightblue') 
v.get_patch_by_id('11').set_color('darkblue') 

v.get_label_by_id('10').set_text(' ') 
v.get_label_by_id('01').set_text(' ') 
v.get_label_by_id('11').set_text(' ') 

v.get_label_by_id('A').set_text('Set A') 
v.get_label_by_id('B').set_text('Set B') 
plt.gca().set_facecolor("white")
plt.gca().set_axis_on()
```
